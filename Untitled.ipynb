```python
import os 
import tensorflow 
from tensorflow import keras 
from keras.preprocessing import image 
import cv2
import matplotlib.pyplot as plt 
```

```python
categores = ['0','1']
```

```python
for categorey in categores : 
    path=os.path.join('mask_images','train',categorey)
    print(path)
    for file in os.listdir(path): 
        print(file)
        path1=os.path.join('mask_images','train',categorey,file)
        k=plt.imread(path1)
        plt.imshow(k)
        plt.show()
        break
    
```

```python
data=[]
for categorey in categores : 
    path=os.path.join('mask_images','train',categorey)
    # print(path)
    lable=categores.index(categorey)
    for file in os.listdir(path): 
        
        imge_path=os.path.join(path,file)
        image=cv2.imread(imge_path)
        if image.shape[0]<224 or image.shape[1]<224 : 
            print(image.shape)
        else : 
            image = cv2.resize(image,(224,224))
            data.append([image,lable])
```

```python
len(data)
```

```python
import random 
random.shuffle(data)
```

```python
X=[]
y=[]

for featues , lable in data :
    X.append(featues)
    y.append(lable)
```

```python
len(X)
```

```python
len(y)
```

```python
import numpy as np 
X = np.array(X)
y=np.array(y)
```

```python
X.shape
```

```python
y.shape
```

```python
X=X/225
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from tensorflow import keras 
from keras.preprocessing import image 
from keras.applications import VGG16 
vgg =VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential 
model = Sequential()
```

```python
for layer in vgg.layers[:-1] : 
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers : 
    layer.trainable = False 
```

```python
model.summary()
```

```python
from keras.layers import Dense 
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam' , loss = 'binary_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
history = model.fit(X_train,y_train,
                    validation_data=(X_test,y_test),
                    epochs=9,  # Continue from epoch 6 to 10
                    initial_epoch=5,  # Starts from epoch 6
                    batch_size=32)
```
